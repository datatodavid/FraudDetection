Only admitted if inpatient, so 'Length of Stay' calculated with Inpatient dataset

```python
Train_Inpatient['Length_of_Stay'] = pd.to_datetime(Train_Inpatient['DischargeDt']) - pd.to_datetime(Train_Inpatient['AdmissionDt'])
Train_Inpatient['Length_of_Stay'] = Train_Inpatient['Length_of_Stay'].astype(str).str.extract('(\d+)').astype(int) + 1

Train_Inpatient['Reimbursement_per_Day'] = Train_Inpatient['InscClaimAmtReimbursed'] / Train_Inpatient['Length_of_Stay']

Train_Inpatient['MismatchedAdmitCode'] = np.where(Train_Inpatient['ClmAdmitDiagnosisCode'] == Train_Inpatient['ClmDiagnosisCode_1'], 0, 1)
```

## Create Master Dataset with Inpatient and Outpatient Visits

```python
Train_All['Length_of_Claim'] = pd.to_datetime(Train_All['ClaimEndDt']) - pd.to_datetime(Train_All['ClaimStartDt'])
Train_All['Length_of_Claim'] = Train_All['Length_of_Claim'].astype(str).str.extract('(\d+)').astype(int) + 1

Train_All['MonthofClaimStartDt'] = pd.to_datetime(Train_All['ClaimStartDt']).dt.month_name()
Train_All['WhichThirdofMonth'] = np.where(pd.to_datetime(Train_All['ClaimStartDt']).dt.day > 10, np.where(pd.to_datetime(Train_All['ClaimStartDt']).dt.day > 20, 3, 2), 1)
```

```python
Docs_List = [col for col in Train_All.columns if 'Physician' in col]
Train_All['Num_Physicians_in_Claim'] = Train_All[Docs_List].count(axis=1)

Claim_Procedure_List = [col for col in Train_All.columns if 'ClmProcedureCode_' in col]
Train_All['Num_ClmProcedureCodes'] = Train_All[Claim_Procedure_List].count(axis=1)

Claim_Diagnosis_List = [col for col in Train_All.columns if 'ClmDiagnosisCode_' in col]
Train_All['Num_ClmDiagnosisCodes'] = Train_All[Claim_Diagnosis_List].count(axis=1)

Train_All['Num_ClmCodes'] = Train_All['Num_ClmProcedureCodes'] + Train_All['Num_ClmDiagnosisCodes']
```

```python
def avg(df):
    return round(df.mean(skipna=True),2)

def med(df):
    return df.median(skipna=True) 

def zeroes(df):
    return np.sum(df==0)

def NAs(df):
    return np.sum(df.isna())

def mode(df):
    try: 
        mode_val = df.mode().values[0]
    except:
        mode_val = np.nan
    return mode_val

def mode_freq(df):
    try:
        mode_val_freq = sum(df==df.mode().values[0])
    except:
        mode_val_freq = np.nan
    return mode_val_freq

def mode2(df):
    try: 
        mode_val2 = df[df!=df.mode().values[0]].mode().values[0]
    except:
        mode_val2 = np.nan
    return mode_val2

def mode_freq2(df):
    try:
        mode_val_freq2 = sum(df==df[df!=df.mode().values[0]].mode().values[0])
    except:
        mode_val_freq2 = np.nan
    return mode_val_freq2

def mode3(df):
    try: 
        mode_val3 = df[(df!=df.mode().values[0]) & (df!=df[df!=df.mode().values[0]].mode().values[0])].mode().values[0]
    except:
        mode_val3 = np.nan
    return mode_val3

def mode_freq3(df):
    try:
        mode_val_freq3 = sum(df==df[(df!=df.mode().values[0]) & (df==df[df!=df.mode().values[0]].mode().values[0])].mode().values[0])
    except:
        mode_val_freq3 = np.nan
    return mode_val_freq3
    
def num_unique(df):
    return len(np.unique(df[~np.isnan(df)]))
#     len(df.unique())
  
```
